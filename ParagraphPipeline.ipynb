importing libraries and downloading important resources


```python
# importing libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

# download required nltk resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

Defining the paragraph.

```python
paragraph = (
   """Technology has become a vital component of contemporary life, changing how people work, communicate, and acquire knowledge. From mobile devices to artificial intelligence, technological advancements have enhanced convenience and productivity across nearly every sector. In the field of education, online learning tools enable students to study from any location, while in healthcare, innovative medical technologies and data-driven insights help save numerous lives. Despite these advantages, the swift expansion of technology brings challenges such as privacy issues, job displacement, and excessive reliance on machines. Thus, although technology provides remarkable opportunities, it is crucial to apply it responsibly to ensure that development contributes to a fairer and more sustainable society."""
)
print(paragraph)
```

Tokenization : This splits the paragraph into individual words and punctuation marks


```python
tokens = word_tokenize(paragraph)
print("Token")
print(tokens)
```

Stopword Removal : This removes common english words like "is","the","and",etc. Keeping only meaningful words.

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("after stopword removal")
print(filtered_tokens)
```

Stemming : Reduce a word to its root forms often not actual words

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("after stemming")
print(stemmed_words)
```

Lemmatization : lemmatization returns dictionary base forms.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("After lemmatization")
print(lemmatized_words)
```
